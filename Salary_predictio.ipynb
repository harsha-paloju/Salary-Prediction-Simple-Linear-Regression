```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("Machine Learning Datasets Updated/19. Simple Linear Regression/Salary_Data.csv")
df
```

```python
df.rename(columns={"YearsExperience":"exp"},inplace=True)
```

```python
df
```

```python
sns.scatterplot(data=df,x="exp",y="Salary")
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(df.iloc[:,[0]],df.iloc[:,1],test_size=0.2,random_state=0)
xtest
```

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg.fit(xtrain,ytrain)

```

```python
reg.predict([[11]])
```

```python
from sklearn.metrics import r2_score
```

```python
r_squared=r2_score(ytest,reg.predict(xtest))
```
